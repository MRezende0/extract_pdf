```python
import os
import glob
import pandas as pd
```

```python
# Diretório onde estão os PDFs renomeados
source_dir = 'pdf_renomeado'

# Diretórios de destino para cada unidade (certifique-se de que essas pastas já existem)
dest_dir_21 = '21'
dest_dir_22 = '22'

# Carrega a planilha do Excel (ajuste o caminho e o nome do arquivo)
excel_path = 'Áreas  Cadastradas_forn.xlsx'
df = pd.read_excel(excel_path, sheet_name=2)
```

```python
# Cria um dicionário mapeando setor para unidade
# Supondo que as colunas da planilha se chamem 'setor' e 'unidade'
mapping = {}
for index, row in df.iterrows():
    # Converte o setor para string e remove espaços em branco
    setor_val = str(row['SETOR']).strip()
    unidade = row['EMPRESA']
    mapping[setor_val] = unidade
```

```python
# Lista todos os PDFs do diretório
pdf_files = glob.glob(os.path.join(source_dir, '*.pdf'))
```

```python
for pdf_file in pdf_files:
    # Obtém o nome do arquivo
    filename = os.path.basename(pdf_file)
    
    # Extrai o setor do nome do arquivo (supondo o formato "setor - SAFRA.pdf")
    setor_from_filename = filename.split(' ')[0].strip()
    
    # Verifica se o setor está presente no mapeamento
    if setor_from_filename in mapping:
        unidade = mapping[setor_from_filename]
        # Define o diretório de destino com base na unidade
        if unidade == 21:
            new_path = os.path.join(dest_dir_21, filename)
        elif unidade == 22:
            new_path = os.path.join(dest_dir_22, filename)
        else:
            print(f"Unidade não reconhecida para o arquivo {filename}: {unidade}")
            continue
        
        # Move o arquivo para a pasta correspondente
        os.rename(pdf_file, new_path)
        print(f"Arquivo {filename} movido para {new_path}")
    else:
        print(f"Setor {setor_from_filename} não encontrado na planilha para o arquivo {filename}")
```
