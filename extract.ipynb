```python
# Defina o diretório onde estão os seus PDFs
pdf_dir = 'Mapas Individuais'
pdf_files = glob.glob(os.path.join(pdf_dir, '*.pdf'))
dest_dir = "pdf_renomeado"
```

```python
for pdf_file in pdf_files:
    try:
        # Abre o PDF e extrai o texto da primeira página
        reader = PdfReader(pdf_file)
        page = reader.pages[1]
        text = page.extract_text()
        
        # Extrai o valor do setor (assumindo que está na primeira linha)
        setor = text.split('\n')[1].split(' ')[0]
        
        # Cria o novo nome de arquivo com o padrão "setor - SAFRA.pdf"
        new_filename = f"{setor} - SAFRA.pdf"
        new_filepath = os.path.join(dest_dir, new_filename)
        
        # Renomeia o arquivo
        os.rename(pdf_file, new_filepath)
        print(f"Arquivo renomeado: {pdf_file} -> {new_filepath}")
        
    except Exception as e:
        print(f"Erro ao processar o arquivo {pdf_file}: {e}")
```
